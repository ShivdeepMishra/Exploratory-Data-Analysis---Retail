## Exploratory Data Analysis

### Density-  Sales Vs Profit

```python
plt.figure(figsize = (8,5))

sns.kdeplot(data.Sales, color = 'yellow', label = 'Sales', shade = True)
sns.kdeplot(data.Profit, color = 'dodgerblue', label = 'Profit', shade = True)

plt.xlim([-100, 1000])
plt.legend()
plt.show()
```

```python
# plotting the relation between Discount and Profit

sns.lineplot(x = 'Discount', y = 'Profit', color = 'r', data = data)
plt.show()
```

```python
# correlating the columns

data.corr()
```

```python
# plotting the correlation using heatmap

plt.figure(figsize=(8,6))
sns.heatmap(data.corr(), annot = True, cmap = "viridis")
plt.show()
```

     - Sales and Profit are moderately correlated.

     - Discount and Profit are negatively correlated.
 
     - Quantity and Sales are less correlated than moderate.

### Bivariate and Univariate distributions

```python
# Category wise

sns.pairplot(data, hue = 'Category', palette = "tab10")
plt.show()
```

```python
# Region wise

sns.pairplot(data, hue = 'Region')
plt.show()
```

```python
# count plot of each column

import warnings
warnings.filterwarnings("ignore")

sns.set_style("whitegrid")

fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 7))

sns.countplot(data['Category'], ax = ax[0][0])
ax[0][0].set_title('Category', fontsize = 20)

sns.countplot(data['Segment'], ax = ax[0][1])
ax[0][1].set_title('Segment', fontsize = 20)

sns.countplot(data['Ship Mode'], ax = ax[1][0])
ax[1][0].set_title('Ship Mode', fontsize = 20)

sns.countplot(data['Region'], ax = ax[1][1])
ax[1][1].set_title('Region', fontsize = 20)

plt.tight_layout()
```

```python
plt.figure(figsize = (20, 10))

sns.countplot(data['Sub-Category'])

plt.title('Sub-Category', fontsize = 20)

plt.show()
```

```python
plt.figure(figsize = (18, 10))

sns.countplot(data['State'])

plt.xticks(rotation = 90)

plt.title('State', fontsize = 20)

plt.show()
```

```python
plt.figure(figsize = (18, 10))

sns.countplot(data['Discount'])

plt.xticks(rotation = 90)

plt.title('Discount', fontsize = 20)

plt.show()
```

```python
# Distribution of the data using the plot
```

```python
fig, ax = plt.subplots(ncols = 2, nrows = 2, figsize = (20, 10))

sns.distplot(data['Sales'], ax = ax[0][0], color = 'dodgerblue')
ax[0][0].set_title('Sales Distribution', fontsize = 20)

sns.distplot(data['Profit'], ax = ax[0][1], color = 'g')
ax[0][1].set_title('Profit Distribution', fontsize = 20)

sns.distplot(data['Quantity'], ax = ax[1][0], color = 'red')
ax[1][0].set_title('Quantity Distribution', fontsize = 20)

sns.distplot(data['Discount'], ax = ax[1][1], color = 'purple')
ax[1][1].set_title('Discount Distribution', fontsize = 20)

plt.show()
```

### State wise Deal

```python
data_state_value = data['State'].value_counts()
data_state_value.head(20)
```

```python
data_state_value.plot(kind = 'pie', figsize = (35, 25), autopct = '%1.1f%%')

plt.ylabel('Frequency / Number of deals', fontsize = 20)

plt.xlabel('State', fontsize = 20)

plt.title('State Wise Deal', fontsize = 20)

plt.show()
```

```python
data['State'].value_counts().mean()
```

### City Wise Deal

```python
data_city_value = data['City'].value_counts()

data_city_value = data_city_value.head(50)

data_city_value.plot(kind = 'pie', figsize = (25, 25), autopct = '%1.1f%%')

plt.ylabel('Frequency / Number of deals', fontsize = 20)

plt.xlabel('City', fontsize = 20)

plt.title('City Wise Deal', fontsize = 20)

plt.show()
```

```python
data.City.value_counts().mean()
```

### Region wise Sales Vs Profits

```python
sales = data.groupby('Region').sum()['Sales'].sort_values(ascending = False)

profits = data.groupby('Region').sum()['Profit'].sort_values(ascending = False)

sales.plot(kind = 'bar', figsize = (10, 5), fontsize = 15, color = 'yellow')
profits.plot(kind = 'bar', figsize = (10, 5), fontsize = 15, color = 'dodgerblue')

plt.xlabel('Regions', fontsize = 20)
plt.ylabel('Total Sales & Profits', fontsize = 20)
plt.title('Region wise Sales vs Profits', fontsize = 20)

plt.legend()
plt.show()
```

### City wise Analysis of the Profit

```python
# City wise Analysis of the Profit

data_city = data.groupby(['City'])[['Sales', 'Discount', 'Profit']].mean().sort_values('Profit')
data_city.head() 
```

```python
data_city['Profit'].head(30).plot(kind = 'bar', figsize = (15, 5), color = 'red')

plt.title("City wise Lowest Profit", fontsize = 20)
plt.show()
```

```python
data_city['Profit'].tail(30).plot(kind = 'bar', figsize = (15, 5), color = 'dodgerblue')

plt.title("City wise Top Profit", fontsize = 20)
plt.show()
```

### State wise Analysis of the Profit

```python
data_state = data.groupby('State')[['Sales', 'Discount', 'Profit']].mean().sort_values('Profit')
data_state.head(10)
```

```python
data_state.Profit.plot(kind = 'bar', figsize = (15, 4))

plt.title('State wise Profit Analysis', fontsize = 20)
plt.ylabel('Profit per State')
plt.xlabel('States')

plt.show()
```

### Category wise Profit Analysis

```python
data_category = data.groupby("Category")[['Sales', 'Discount', 'Profit']].mean()

data_category
```

```python
data_category.plot.pie(subplots = True,
                      figsize = (18, 20),
                      autopct = "%1.1f%%",
                      labels = data_category,
                       cmap = "Set2")

plt.show()
```

### Sub-Category wise Analysis

```python
data_sub_category_loss = data.groupby('Sub-Category').min()['Profit'].sort_values()
```

```python
plt.figure(figsize = (8, 6))

data_sub_category_loss.plot.bar(color = 'red')

plt.title('Sub-Categories with highest loss')
plt.xticks(rotation = 60, fontsize = 15)
plt.yticks(fontsize = 15)

plt.show()
```

```python
# Sub-Category wise Sales, Profit and Discount

data_sub_category = data.groupby(['Sub-Category'])[['Sales', 'Discount', 'Profit']].mean()

data_sub_category.head(10)
```

```python
plt.figure(figsize = (35, 15))

plt.pie(data_sub_category.Sales, labels = data_sub_category.index, autopct = '%1.1f%%')

plt.title('Sales Analysis (Sub-Category Wise)', fontsize = 20)
plt.legend()
plt.xticks(rotation = 90)

plt.show()
```

```python
plt.figure(figsize = (35, 15))

plt.pie(data_sub_category.Discount, labels = data_sub_category.index, autopct = '%1.1f%%')

plt.title('Discount Analysis (Sub-Category Wise)', fontsize = 20)
plt.legend()
plt.xticks(rotation = 90)

plt.show()
```

## Conclusion:

- In some cases, Profit Percentage is more than sales and vice-versa.

- Sales and Profit are moderately correlated.

- Discount and Profit are strongly negatively correlated.

- The lowest amount of sales = South Dakota (0.8%)

- California, New York, Texas are the three states with the highest deal.

- Ohio has the lowest profit and Vermont has the highest profit.

- Sales are high for the same-day ships.

- New York City has the highest city-wise deals.

West region shows more sales and more profit whereas the central region shows the least sales and profits. So, one should try to increase sales and profits in the central region. Profit percentage can be increased in some cases than sales. The loss in subcategories can be converted into profit percentage.
